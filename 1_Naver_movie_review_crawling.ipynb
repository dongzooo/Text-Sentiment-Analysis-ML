## 댓글 자동수집화 함수

```python
#댓글 크롤링함수 , 요청요류가나도 text_reple속에 데이터는 존재
def get_reple(page=1):
    response = requests.get(url.format(page))
    soup = BeautifulSoup(response.text, 'html.parser')
    
    text_reple = []
    
    for li in soup.find('div', {'class':'score_result'}).find_all('li'):
#         print('댓글: ', li.p.text)
        text_reple.append(li.p.text)
    return text_reple
```

```python
# \n \t 등 불필요한 텍스트 제거 함수
def get_cleantext(text):
    cleaned_text = re.sub('[a-zA-Z]' , '', str(text))
    cleaned_text = re.sub('[\{\}\[\]\/?.,;:|\)*~`!^\-_+<>@\#$%&\\\=\(\'\"]','', str(cleaned_text))
#     cleaned_text = re.sub('관람객','', str(cleaned_text))
    return cleaned_text
```

# 크롤링 시작

```python
#크롤링 자동화 반복문
temp_text = []
for i in range(1,300):
    time.sleep(0.5)
    print('요청횟수',i, end = '\r')
    t = get_reple(i)
    temp_text += t
```

```python
text_data = get_cleantext(temp_text)
```
